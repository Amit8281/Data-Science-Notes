## Encoder and Imputer

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()
```

```python
df2 = lab_enc.fit_transform(df['City'])
pd.Series(df2)
```

```python
df['City'] = df2
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
si = SimpleImputer()
```

```python
import pandas as pd
df=pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,55000],
                'City':['Bengaluru','Delhi','Hydrabad','Bengaluru','Hydrabad','Bengaluru'],
                'Gender':['Male','Female','Female','Female','male','Male'],
                'Exp':[1,3,5,6,9, None]})

df
```

```python
ct = make_column_transformer(
    (ohe,['City','Gender']),
    (si,['Exp']),
    remainder = 'passthrough') # 'passthrough to keep all other columns'
```

```python
encoded = pd.DataFrame(ct.fit_transform(df))
encoded
```

```python
encode = pd.DataFrame(ct.fit_transform(df),columns=['City_Bengaluru','City_Delhi','City_Hyd','Gender_Female','Gender_male','Exp','exp','Salary'])
```
