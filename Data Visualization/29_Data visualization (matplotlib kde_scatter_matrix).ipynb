# (matplotlib kde_scatter_matrix)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.plotting import scatter_matrix
```

```python
#Load dataset from seaborn
mileage = sns.load_dataset('mpg')
mileage.head()
```

```python
#with two columns
plt.scatter(x=mileage['cylinders'],y=mileage['mpg'])
plt.xlabel('Cyl')
plt.ylabel('mpg')
plt.show()
```

```python
#Lets plot scatter_matrix to check relation of all column's
```

```python
#scatter_Matrix
scatter_matrix(mileage, figsize= (15,10), color='b')
plt.show()
```
