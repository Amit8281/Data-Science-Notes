# Catagorical Data Plot

**~catplot**

**~stripplot**

**~swarmplot**

**~countplot**

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#catplot()
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.countplot(x='day',data=tips)
plt.yticks(range(10,100,10))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.grid()
sns.countplot(x='day',hue='time',data=tips)
plt.yticks(range(0,110,10))
plt.show()
```

**similarly you can practice for other catagories**

```python
sns.stripplot(x='size',y='total_bill',data=tips)
plt.show()
```

```python
sns.stripplot(x='smoker',y='total_bill',data=tips)
plt.show()
```
