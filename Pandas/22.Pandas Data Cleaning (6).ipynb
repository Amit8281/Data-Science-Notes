**Operating on Null Values**

As we have seen pandas have treat None as NaN as essentially interchargeble for indicating missing or null value . To facilitate this convention, there are several useful method for detecting removing and replacing null values in pandas data structures: they are:--

. isnull(): Generate a boolean mask indicating missing values

notnull() : opposite of isnull ()

dropna() : return a filtered version of data

fillna(): return a copy of the data with missing values filled or imputed

We will conclude this session with a brief exploration and demonstration of these routines

# Create Dataframe with missing values

```python
raw_data = {'first_name': ['Jason',np.nan,'Tina','Jackl','Amy','madhu'],
           'last_name': ['Miller',np.nan,'Ali','Milner','Cooze','Patil'],
           'age': [42,np.nan,36,24,73,25],
           'sex' : ['M',np.nan,'F','M','F','F'],
           'unit-1': [4,np.nan, np.nan, 2, 3, 4],
           'unit-2' : [3, np.nan, np.nan, 4, 3.5, 4]}
df = pd.DataFrame(raw_data)
df
```

**Finding nulls (NaN's) in all the columns**

```python
# It is diffrent to count nulls if the DataFrame is big. Using isnull().sum() we can find it easily
df.isnull().sum()
```

```python
#Another way
df.isna().sum()
```

```python
#notnull shows how many rows do not have nulls( opposite of nulls)
df.notnull().sum()
```

**Drop missing observation**

```python
df_no_missing = df.dropna()
df_no_missing
```

**Drop rows where all cellls in that row is NA**

```python
df
```

```python
#row's that has all NAN's will be removed
df_cleaned = df.dropna(how='all')
df_cleaned
```

**Creat a new column full of missing values**

```python
df['Final_Score'] = np.nan
df
```

**Drop column if they only contain missing values**

```python
dd =df.dropna(axis=1, how='all')
dd
```

```python
df
```

**Fill in missing in Unit-1 with the mean value of Unit-1**

inplace = True means that the changes are saved to the df right away

```python
df["unit-1"].fillna(df["unit-1"].mean(),inplace=True)
df
```

**Fill in missing in Unit-2 with each sex mean value of Unit-2**

```python
df["unit-2"].fillna(df.groupby("sex")["unit-2"].transform("mean"), inplace = True)
df
```

```python
#back_fill
df.fillna(method='bfill')
```

```python
#forward_fill
df.fillna(method = 'ffill', inplace = True)
```

```python
df
```

```python
# you can add new rows by location(index) as below
df.loc[6] = ['Harry','Kene',28,'M',3.5,3.5,np.nan]
df
```

```python
df['Final_Score'].fillna(df['unit-1'] + df['unit-2'], inplace=True)
df
```

```python
#you can modify the entire row by locating the index number and providing complete details
df.loc[1] = ['Rohit','Sharma',35,'M',4.5,4,8.5]
df
```

**Hierachical Indexing**

**Creat DataFrame**

```python
raw_data = {'Company': ['Gooogle','Gooogle','Gooogle','Gooogle','Facebook','Facebook','Facebook','Facebook','Amazon','Amazon','Amazon','Amazon',],
           'Project': ['1st','1st','2nd','2nd','1st','1st','2nd','2nd','1st','1st','2nd','2nd'],
           'Incharge': ['Sunder','Ruth','Benjamin','Sergey','Mark','Shery','Edurdo','David','Jeffery','Andy','Keith','Edith'],
           'Sales ( in millions)': [54,24,31,12,23,24,34,31,32,23,52,13],
           'Profit (in millions)': [14,8,9,4,10,11,14,13,12,7,22,4]}
df = pd.DataFrame(raw_data)
df
```

**Set the hierarchical index but leave the columns inplace**

```python
df.set_index(['Company', 'Project','Incharge'],drop=False)
```

**Set the hierarchical index to be by Company, and then by Project**

```python
df.set_index(['Company', 'Project','Incharge'],drop = True , inplace = True)
df
```

```python
df
```

**Swap the levels in the index**

```python
df.swaplevel('Company','Project')
```

**Summarize the result by Company**

```python
df.sum(level = 'Company')
```

```python
df.sum(level= 'Project')
```

```python
df.mean(level = 'Company')
```
